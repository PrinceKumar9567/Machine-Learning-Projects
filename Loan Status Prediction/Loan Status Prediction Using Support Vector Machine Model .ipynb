Importing Dependencies

```python
import numpy as np 
import pandas as pd
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import LabelEncoder
```

```python
#Data Collection and Processing
data = pd.read_csv(url)
#printing first five rows of the data
data.head()
```

```python
#Checking for shape of the data
data.shape
```

```python
#Statistical measures of the data
data.describe()
```

```python
#Checking for missing values
data.isnull().sum()
```

```python
#Deleting all the rows in which missing data is present
data.dropna(axis=0,inplace=True)
```

```python
data.shape
#Printing all the columns of the data
data.columns
```

```python
#Converting all the categorical data into numerical ones using labelencoder
le = LabelEncoder()
for i in ['Gender','Married','Education','Self_Employed','Property_Area','Loan_Status','Loan_ID']:
    data[i] = le.fit_transform(data[i])
data.head()    
```

```python
# Counting the value 3+ in Dependencies column 
red_count = (data['Dependents'] == '3+').sum()
print(red_count)

# Replacing all "3+" with 4
# Use the .loc accessor to properly update values in the dataframe
data.loc[data['Dependents'] == "3+", 'Dependents'] = 4

# Alternative method using replace
# data['Dependents'] = data['Dependents'].replace("3+", 4)
```

As we can see this correlation data some columns like education,loanAmount,Applicantincome these are not contributing in the loan approval so we can drop these columns we will not going to consider because it is just the id  


```python
#Splitting the data for target values
X = data.drop(columns = ['Loan_ID','Loan_Status'])
y = data['Loan_Status']  
```

```python
#Splitting the data for testing and training
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.1,stratify = y,random_state = 2)
print(X.shape,X_train.shape,X_test.shape)
```

```python
#Training the model
classifier = svm.SVC(kernel = 'linear')
classifier.fit(X_train,y_train)
```

Model Evaluation

```python
#Evaluating the ata on training data
X_train_pred = classifier.predict(X_train)
accuracy_train = accuracy_score(X_train_pred,y_train)
print("Accuracy of the model on the training data is:", accuracy_train)
```

```python
#Evaluating the model on testing data
X_test_pred = classifier.predict(X_test)
Accuracy_test = accuracy_score(X_test_pred,y_test)
print("Accuracy of the model on the testing data is:", Accuracy_test)
```
