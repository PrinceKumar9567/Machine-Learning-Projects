```python
url = r"C:\Users\princ\Downloads\Chapter 1\Python\Machine Learning\ML Data\creditcard.csv.zip"
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
#loading the dataset to a pandas data frame
Data = pd.read_csv(url)
CC_data = pd.DataFrame(Data)
```

```python
#printing 1st five rows of data--(Class = 0-Normal transaction,Class = 1-Fradulent Transaction)
CC_data.head()
```

```python
#checking for missing values
CC_data.isnull().sum()
```

```python
#Dataset information
CC_data.info()
```

```python
#Distribution of legit and fraud transaction
CC_data['Class'].value_counts()
```

This is a highly unbalanced dataset


```python
#Separeting the data for analysis
legit = CC_data[CC_data.Class == 0]
fraud = CC_data[CC_data.Class == 1]
```

```python
print(legit.shape)
print(fraud.shape)
```

```python
#statistical measures of data
legit.Amount.describe()
```

```python
 fraud.Amount.describe()
```

```python
#comparing the values of legit and fraud 
CC_data.groupby('Class').mean()
```

```python
#Building a sample dataset for similar distribution of both the transactions
legit_sample = legit.sample(n = 492)
```

```python
#concatenating two dataframe
new_data = pd.concat([legit_sample,fraud],axis = 0)
```

```python
new_data.head()
```

```python
new_data['Class'].value_counts()
```

```python
new_data.groupby('Class').mean()
```

```python
#Spliting the data
X = new_data.drop(columns = 'Class',axis = 1)
y = new_data['Class']
```

```python
#Splitting the data further for training and testing
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.2,stratify = y,random_state=2)
```

```python
# training the model
model = LogisticRegression()
model.fit(X_train,y_train)
```

Evaluation of the model


```python
X_train_pred = model.predict(X_train)
train_accuracy = accuracy_score(X_train_pred,y_train)
print(f"Accuracy on Training Data is: {train_accuracy}")
```

```python
X_test_pred = model.predict(X_test)
test_accuracy = accuracy_score(X_test_pred,y_test)
print(f"Accuracy Score on Testing Data is: {test_accuracy}")
```
